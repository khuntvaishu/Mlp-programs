```python
sns.scatterplot(x="mass",y="distance",data=planets)
```

```python
sns.scatterplot(x="mass",y="distance",data=planets, hue="year")
```

```python
sns.scatterplot(x="mass",y="distance",data=planets, hue="year",size="mass")
```

```python
sns.scatterplot(x="mass",y="distance",data=planets, hue="year",size="mass",palette='YlGnBu')
```

```python
sns.histplot(planets['year'],kde=True,bins=15)#kde=line,bins=boxes
```

```python
sns.displot(planets['year'])
```

```python
sns.barplot(x='year',y='mass',data=planets,palette='YlGnBu')
```

```python
sns.boxplot(x='year',y='mass',data=planets,palette='YlGnBu')
```

```python
sns.stripplot(x='number',y='mass',data=planets, hue='year',palette='YlGnBu',dodge=True)
```

```python
sns.jointplot(x='number',y='mass',data=planets)
```

```python
sns.jointplot(x='number',y='mass',data=planets,kind="reg")
```

```python
sns.jointplot(x='number',y='mass',data=planets,kind="kde")
```
